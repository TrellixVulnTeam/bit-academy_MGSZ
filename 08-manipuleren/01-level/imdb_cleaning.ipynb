Dubbele datapunten verwijderen uit een dataframe.

Import libraries.

```python
import pandas as pd
```

Get csv file.

```python
df = pd.read_csv('IMDb_films.csv', index_col='index')
df
```

```python
len(df)
print("Oorspronkelijke datapunten:", len(df))
```

Delete duplicates.

```python
df.drop_duplicates()
```

```python
print("Duplicates:", len(df)-len(df.drop_duplicates()))
```
