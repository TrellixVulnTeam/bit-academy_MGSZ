# Scoren bij sjeik Pan al Andas

We installeren de pandas package.

```python
import pandas as pd
```

We halen CSV file op.

```python
df = pd.read_csv("fifa_ratings.csv")
```

We printen dataset.

```python
df
```

We halen de kolommen club en value_eur op. Grouperen de clubs en rangschikken van hoog naar laag de top 10 clubs.

```python
df1 = df[['club', 'value_eur']].groupby('club').sum().sort_values(by=['value_eur'], ascending=False).head(10)
```

We printen dataset.

```python
df1
```

We roepen matplotlib inline op.

```python
%matplotlib inline
```

We plotten de gerangschikte clubs in een plot.

```python
df1 = df[['club', 'value_eur']].groupby('club').sum().sort_values(by=['value_eur'], ascending=False).head(10).plot(kind='bar')
```
