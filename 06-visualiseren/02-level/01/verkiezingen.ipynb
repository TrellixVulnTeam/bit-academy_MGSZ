# Lees je data

Om te beginnen lezen we de database uit. Hiervoor hebben we de module csv nodig:

```python
import csv
```

Vervolgens voeren we code uit om het CSV-bestand te laden en op te slaan in de variabele data, opdat we die kunnen gebruiken voor de volgende exercise:

```python
data = [] #creating container for data

with open("president_county_candidate.csv") as csv_file: #open the file
    csv_data = csv.reader(csv_file)  #read the data

    for row in csv_data: #loop through each row
        data.append(row) #append data to container

csv_file.close() #close the file
```

Als de bovenstaande code werkt, dan zou de onderstaande code moeten werken.

```python
data                                  
```

## Heeft iedereen gestemd?

We controleren of iedereen heeft gestemd! Als het goed is, dan zijn er 51 staten in totaal. Dit zijn de 50 staten van de Verenigde Staten plus Washington D.C., wat als in deze database als een staat wordt beschouwd.

```python
staten = []
for row in data[1:]:
    staat = row[0]
    staten.append(staat)
    
aantal_staten = (len(set(staten)))

print("De database bevat", aantal_staten, "staten")
 
```

## Wie heeft er gewonnen?

We kijken wie er gewonnen heeft. Joe Biden of Donald Trump?

```python
trump = 0
biden = 0

for i in data[1:]:
    if i[2] == "Donald Trump":
        trump = trump + int(i[4])
    elif i[2] == "Joe Biden":
        biden = biden + int(i[4])
    else:
        pass
    
if biden > trump:
    print("Joe Biden heeft gewonnen!")
elif biden < trump:
    print("Donald Trump heeft gewonnen!")
else:
    print("De stemmen zijn gelijk!")
```
